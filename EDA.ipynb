```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=sns.load_dataset("iris")
print(data)
```

```python
from sklearn.preprocessing import LabelEncoder
v=LabelEncoder()
x=['sepal_length','petal_length','species']
for i in x:
    data[i]=v.fit_transform(data[i])
data
```

```python
from sklearn.preprocessing import StandardScaler
k=StandardScaler()
data[['species']]=k.fit_transform(data[['species']])
print(data)
```

```python
corelation_matrix=data.corr()
print(corelation_matrix)
```

```python
sns.heatmap(corelation_matrix,annot=True)
```

```python
data=sns.load_dataset("penguins")
print(data)
```
